```python
import redis
import json
from redis.commands.json.path import Path
```

```python
with open("large-file.json") as f:
    data = json.load(f)
```

```python
data[0]
```

```python
r = redis.Redis(host="127.0.0.1", port=6379, db=0)
```

```python
str_mapping_keys = []
for key in data[0].keys():
    if isinstance(data[0][key], str):
        str_mapping_keys.append(key)

str_mapping_keys
```

```python
dict_mapping_keys = []
for key in data[0].keys():
    if isinstance(data[0][key], dict):
        dict_mapping_keys.append(key)
```

```python
str_data = []
for i in range(len(data)):
    cur_elem = {}
    for key in str_mapping_keys:
        cur_elem[key] = data[i][key]
    str_data.append(cur_elem)
```

```python
int_data = []
for i in range(len(data)):
    cur_elem = {}
    cur_elem["id"] = data[i]["id"]
    int_data.append(cur_elem)
```

### hset init, hset and hget

```python
[r.hset(f"id:{i}", mapping=str_data[i]) for i in range(len(str_data))]
```

```python
new_value = {}
for key in str_mapping_keys:
    new_value[key] = "abrakadabra"
r.hset(f"id:{len(str_data)}", mapping=new_value)
```

```python
r.hget(name=f"id:{len(str_data)}", key=str_mapping_keys[1])
```

### zset init, zset and zget

```python
[r.zadd(f"id{i}", mapping=int_data[i]) for i in range(len(int_data))]
```

```python
r.zadd(f"id{len(int_data)}", mapping={"id": 123456789})
```

```python
r.zcount(f"id{len(int_data)}", 1234, 1234567890)
```

### list

```python
id_data = [{f"id{i}": data[i]["id"]} for i in range(len(data))]
```

```python
[r.set(f"id{i}", id_data[i][f"id{i}"]) for i in range(len(id_data))]
```
